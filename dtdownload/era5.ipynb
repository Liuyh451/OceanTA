```python
import cdsapi
import os

# 设置下载路径
save_dir = r"E:\Dataset\ERA5\waves"
os.makedirs(save_dir, exist_ok=True)
save_path = os.path.join(save_dir, "era5_wave_data_2021.nc")

# 下载配置
dataset = "reanalysis-era5-single-levels"
request = {
    "product_type": "reanalysis",
    "variable": [
        "10m_u_component_of_wind",
        "10m_v_component_of_wind",
        "mean_wave_direction",
        "mean_wave_period",
        "significant_height_of_total_swell"
    ],
    "year": "2021",
    "month": [
        "01", "02", "03",
        "04", "05", "06",
        "07", "08"
    ],
    "day": [f"{i:02d}" for i in range(1, 32)],
    "time": [f"{i:02d}:00" for i in range(24)],
    "format": "netcdf",  # 用 format 而不是 data_format 或 download_format
    "area": [25, 120, 15, 130]  # 北，西，南，东 (lat/lon)
}

# 创建客户端并下载
client = cdsapi.Client()
client.retrieve(dataset, request).download(save_path)
```
